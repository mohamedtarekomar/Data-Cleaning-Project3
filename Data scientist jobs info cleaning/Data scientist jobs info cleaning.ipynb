```python
# Importing the data and the libraries

import pandas as pd
pd.set_option("display.max_columns",None)
df = pd.read_csv("Data scientist jobs info cleaning.csv")
df.sample(10)
```

# The things i do to make the data easy to be analyzed :-

## I WILL :-
- split the "Salary Estimate" column to 2 columns min and max and convert them to numerical.
- clean the "Company Name" column.
- clean the "Location" and "Headquarters"columns.

```python
# Take a look at the "Salary Estimate" column's values

df["Salary Estimate"].unique()
```

```python
# Delete all after the " ( " because we do not need it

df['Salary Estimate'] = df["Salary Estimate"].str.split(" ", expand = True)[0]
df['Salary Estimate'] = df["Salary Estimate"].str.split("(", expand = True)[0]

# Delete the dollar sign
df["Salary Estimate"] = df["Salary Estimate"].str.replace("$","")

# Delete the "K" and replace it by "000" and the values that have not "K" are the salary in hour we will handle it..
df["Salary Estimate"] = df["Salary Estimate"].str.replace("K","000")
```

```python
df["Salary Estimate"].unique()
```

```python
# Split the column into 2 columns to make them numerical

df[["Min Salary","Max Salary"]] = df["Salary Estimate"].str.split("-", expand=True)
```

```python
# Delete the useless values

df["Min Salary"] = df["Min Salary"].str.replace("Employer","NaN")
```

```python
df["Max Salary"].unique()
```

```python
# Convert the vvalues into numerical

df["Max Salary"] = df["Max Salary"].astype(float)
df["Min Salary"] = df["Min Salary"].astype(float)
```

```python
# Converting the salary in hour to salary in year to be easy to analyzed


def update_value(value):
    # All the salary in year values are less than 100
    if value < 100:
        return value * 8 * 356
    return value

# Apply the function to the 'Min Salary' column
df['Min Salary'] = df['Min Salary'].apply(update_value)
```

```python
# Converting the salary in hour to salary in year to be easy to analyzed


def update_value(value):
    # All the salary in year values are less than 100
    if value < 100:
        return value * 8 * 356
    return value

# Apply the function to the 'Max Salary' column
df['Max Salary'] = df['Max Salary'].apply(update_value)
```

```python
# Drop the "Salary Estimate" column

df = df.drop(columns="Salary Estimate")
```

```python
df["Company Name"].unique()
```

```python
# Clean the "Company Name" column

df['Company Name'] = df["Company Name"].str.split("\n", expand = True)[0]
```

```python
# The "Size" column has few values and it's better to leave it as it is

df["Size"].unique()
```

```python
df["Location"].unique()
```

```python
# There is a column called "Job Location" that has a values of the location state so we can remove the state from "Location" column

df["Location"] = df["Location"].str.split(", ", expand=True)[0]
```

```python
df["Location"].unique()
```

```python
# Split the state into other column 

df[['Headquarters', 'Headquarters State']] = df['Headquarters'].str.split(', ', n=1, expand=True)
```

```python
# The "Revenue" column has few values and it's better to leave it as it is

df["Revenue"].unique()
```

```python
# Replacing all unnknown values into NaN to be easy to analyzed

df["Revenue"] = df["Revenue"].str.replace("Unknown / Non-Applicable","NaN")
```

```python
# Move the index of the column to there right place to be easy to read

value = df.pop("Min Salary")
df.insert(2,"Min Salary",value)
value = df.pop("Max Salary")
df.insert(3,"Max Salary",value)
value = df.pop("Headquarters State")
df.insert(10,"Headquarters State",value)
value = df.pop("Job Location")
df.insert(8,"Job Location",value)
```
